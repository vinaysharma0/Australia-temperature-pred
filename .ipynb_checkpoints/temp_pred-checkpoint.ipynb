```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('daily-min-temperatures.csv')
```

```python
dataset.head(5)
```

```python
dataset.shape
```

```python
training_set = dataset.iloc[:,1:2].values
```

```python
# feature_scailing
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
training_set = sc.fit_transform(training_set)
```

```python
# diff is 1
X_train = training_set[0:3649]
Y_train = training_set[1:3650]
```

```python
print("X_train shape : {}".format(X_train.shape))
print("Y_train shape : {}".format(Y_train.shape))
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X_train, Y_train, test_size = 0.2)
```

```python
test_X.shape
```

```python
train_X = np.reshape(train_X, (train_X.shape[0], train_X.shape[1], 1))
train_Y = np.reshape(train_Y, (train_Y.shape[0], train_Y.shape[1], 1))
```

```python
test_X = np.reshape(test_X, (test_X.shape[0], test_X.shape[1], 1))
test_Y = np.reshape(test_Y, (test_Y.shape[0], test_Y.shape[1], 1))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, GRU
```

```python
print("train_X shape : {}".format(train_X.shape))
print("train_Y shape : {}".format(train_Y.shape))
```

```python
?LSTM
```

```python
model = Sequential()
model.add(LSTM(units = 20, input_shape=(1, 1)))
model.add(Dropout(0.02))

# model.add(LSTM(units = 20))
# model.add(Dropout(0.1))

model.add(Dense(units = 1))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])


```

```python
history = model.fit(train_X, train_Y, batch_size = 128, validation_data=(test_X, test_Y),shuffle=False, epochs = 300)
```

```python
test_dataset = dataset[:300]
test_dataset.head(5)
```

```python
test_real = test_dataset.iloc[:,1:2].values
```

```python
inputs = test_real
inputs = sc.transform(inputs)
```

```python
inputs = np.reshape(inputs, (300, 1,1))
```

```python
inputs.dtype
```

```python
pred = model.predict(inputs)
```

```python
pred = sc.inverse_transform(pred)
```

```python
pred[:15]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(test_real, color='red', label='real_temperature')
plt.plot(pred, color='blue', label='predicted_temperature')
plt.xlabel('Time')
plt.ylabel('temp')
plt.legend()
plt.show()
```

```python
# plot train and validation loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model train vs validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'])
plt.show()
```
